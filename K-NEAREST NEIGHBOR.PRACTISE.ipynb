```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier(n_neighbors=4)
```

```python
data=pd.read_csv(r"C:\Users\Shivam N Nagpure\Downloads\07-10-2022\07-10-2022\diabetes.csv")
```

```python
data
```

```python
x=pd.DataFrame(data.drop("label",axis=1))
```

```python
y=pd.DataFrame(data["label"])
```

# Training And Testing the Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
KNN.fit(x_train,y_train)
```

```python
x_test
```

```python
y_predict=KNN.predict(x_test)
```

```python
y_predict
```

```python
x_test
```

```python
pd.DataFrame(y_predict)
```

# Evaluating Model

```python
confusion_matrix=metrics.confusion_matrix(y_test,y_predict)
```

```python
confusion_matrix
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.xlabel("actual_value")
plt.ylabel("predicted_vaue")
sns.heatmap(confusion_matrix,annot=True)
plt.show()

```

```python
confusion_matrix=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix)
confusion_matrix.plot()
plt.show()
```
